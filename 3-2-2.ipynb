## 1. 인공지능으로 어떤 문제를 해결할까?

### **(1) 전통적인 문제 해결 vs 인공지능을 적용한 문제 해결**
우리가 일상생활 속에서 접하는 수많은 문제들은 해결 방식에 따라 전통적인 프로그래밍 방식과 인공지능 기계학습 방식으로 나뉩니다. 인공지능은 규칙이 고정되지 않고 인간의 인식, 추론, 학습 형태가 필요한 문제를 해결하는 데 적합합니다.

* **기존 기능 (전통적 프로그래밍)**
  * **스피커**: 전기 신호를 받아 단순 소리 출력
  * **네비게이션**: 고정된 수치 지도 데이터에 기반한 목적지 이동 경로 안내
* **인공지능 기술 적용**
  * **스피커**: 사람의 음성을 인식하고 맥락을 알아들어 대화 가능 및 관련 내용 자동 검색
  * **네비게이션**: 실시간 교통량 데이터를 다각도로 분석하여 최적의 동적 경로 추천

### **(2) 인공지능 없이도 해결 가능한 문제**
해결하는 규칙이 명확하거나 고정되어 있는 유형은 단순 알고리즘식 계산 프로그램만으로 충분히 해결이 가능합니다.
* **단순 계산**: 전화 요금 계산, 놀이공원 입장료 계산, 비만도(BMI) 계산, 세금 정산 등
* **단순 지수 출력**: 수식에 따른 불쾌지수 계산, 센서 측정값 기반 미세먼지 수치 출력 등
* **무작위 선택**: 발표 순서 무작위 정하기, 행운의 번호 추첨 등

### **(3) 인공지능을 적용하면 좋은 문제**
한 가지 고정된 규칙을 적용하기 어렵거나, 데이터의 변화에 따라 실시간으로 규칙을 새롭게 학습하고 갱신해야 하는 도메인에 해당합니다.
* **스팸 메일 분류**: 지속적으로 변형되는 광고성 메일 텍스트 속성을 학습하여 분류
* **자율 주행 자동차**: 실시간으로 변하는 도로 위의 객체(보행자, 차량, 표지판) 시각 인식 및 추론
* **기타 도메인**: 질병 진단 의료 데이터 분석, 텍스트 기반 감정 인식 및 영화 리뷰 긍부정 분류, 과일 등급 자동 시각 분류 등

## 2. 기계학습 모델 구현 5단계 파이프라인

### **단계별 핵심 설계 개념**
1. **문제 정의**: 해결하고자 하는 비즈니스나 연구 목적을 인공지능 문제(분류 또는 회귀) 형태로 명확히 정의하는 단계입니다.
2. **데이터 수집**: 모델 학습 및 검증에 필요한 로우 데이터를 확보하는 단계입니다. 데이터의 대표성과 양(Scale)이 모델의 질을 결정합니다.
3. **데이터 탐색**: 데이터의 구조, 속성(Feature), 결측치 유무를 파악하고 전처리를 수행합니다. 이 단계에서 데이터를 **훈련 데이터**와 **테스트 데이터**로 분리하는 공정이 수행됩니다.
4. **모델링 및 성능 평가**: 선택한 알고리즘에 훈련 데이터를 입력하여 최적의 가중치를 학습시키고, 테스트 데이터를 사용하여 모델의 일반화 성능을 정량적으로 평가합니다.
5. **모델 활용**: 성능이 검증된 최종 완성 모델을 실제 인공지능 시스템이나 스캐너 장비 등에 탑재하여 새로운 데이터에 대한 분류/예측 서비스를 수행합니다.

## 3. 데이터 속성의 유형: 훈련 데이터 vs 테스트 데이터

기계학습 모델을 구축할 때 수집된 전체 데이터 세트는 반드시 목적에 따라 분리되어야 합니다.



### **(1) 훈련 데이터 (Training Data)**
* **역할**: 기계학습 모델을 학습시키고 최적의 규칙(가중치와 편향)을 찾아내는 데 직접 사용되는 데이터 속성 집합입니다.
* **특징**: 모델은 이 데이터의 입력 속성과 정답 레이블 간의 상관관계를 탐색하며 인지 구조를 발달시킵니다.

### **(2) 테스트 데이터 (Test Data)**
* **역할**: 학습을 마치고 완성된 기계학습 모델의 실제 예측 성능 및 일반화 능력을 독립적으로 검증하기 위해 사용하는 데이터 속성 집합입니다.
* **주의사항**: **테스트 데이터는 모델 학습 과정에 절대 관여해서는 안 됩니다.** 시험 문제를 미리 보고 시험을 치는 오류(Data Leakage)를 방지하기 위함입니다. 일반적으로 수집된 전체 데이터의 70%~80%는 훈련 데이터로, 나머지 20%~30%는 테스트 데이터로 분할합니다.

### **(3) 모델 성능의 불균형성: 과대적합과 과소적합**
* **과소적합 (Underfitting)**: 모델의 구조가 지나치게 단순하거나 학습량이 부족하여 훈련 데이터와 테스트 데이터 모두에서 낮은 성능을 보이는 상태입니다.
* **과대적합 (Overfitting)**: 모델이 **훈련 데이터에만 과도하게 맞춰져** 훈련 데이터에 대한 예측률은 매우 높으나, 새로운 데이터인 **테스트 데이터에 대해서는 오차가 크게 발생**하는 상태입니다. 기계학습 모델 구현 시 이를 제어하는 것이 핵심 과제입니다.

```python
# [실습 1] 구글 코랩 환경 설정 및 시각화용 한글 폰트 매핑 수립
import sys
import os
import subprocess
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np

def set_korean_font():
    is_colab = "google.colab" in sys.modules

    if is_colab:
        font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"

        if not os.path.exists(font_path):
            subprocess.run(["apt-get", "update", "-qq"], check=False)
            subprocess.run(["apt-get", "install", "-y", "fonts-nanum"], check=False)
            subprocess.run(["fc-cache", "-fv"], check=False)

        fm.fontManager.addfont(font_path)
        plt.rcParams["font.family"] = "NanumGothic"

    elif sys.platform == "darwin":
        plt.rcParams["font.family"] = "AppleGothic"

    elif sys.platform.startswith("win"):
        plt.rcParams["font.family"] = "Malgun Gothic"

    else:
        plt.rcParams["font.family"] = "NanumGothic"

    plt.rcParams["axes.unicode_minus"] = False

set_korean_font()
print("구글 코랩 실습을 위한 국문 폰트 환경 설정 최적화 완료.")
```

```python
# [실습 2] 전체 데이터의 수집, 탐색 및 훈련/테스트 데이터 세트 분할 구현
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# 1. 교재의 기계학습 데이터 구조를 시뮬레이션하기 위한 가상 속성 데이터 생성 (200개 샘플, 2개 독립속성)
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, 
                           n_redundant=0, n_clusters_per_class=1, random_state=25)

# 2. 3단계 데이터 탐색 과정의 핵심인 데이터 분할 수행 (훈련 데이터 75%, 테스트 데이터 25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=25, stratify=y)

print(f"=== 데이터 속성 및 유형 분할 결과 ===")
print(f"전체 수집된 데이터 크기   : {X.shape}")
print(f"학습용 훈련 데이터 크기   : {X_train.shape} (레이블 수: {y_train.shape[0]})")
print(f"검증용 테스트 데이터 크기 : {X_test.shape}  (레이블 수: {y_test.shape[0]})")

# 3. 분할된 데이터 속성 분포 시각화
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.scatter(X_train[y_train == 0, 0], X_train[y_train == 0, 1], color='#4285F4', marker='o', alpha=0.7, label='정상 (0)')
ax1.scatter(X_train[y_train == 1, 0], X_train[y_train == 1, 1], color='#EA4335', marker='s', alpha=0.7, label='위험/스팸 (1)')
ax1.set_title("훈련 데이터 (Training Data) 속성 분포")
ax1.set_xlabel("속성 1")
ax1.set_ylabel("속성 2")
ax1.legend()
ax1.grid(True, linestyle=':')

ax2.scatter(X_test[y_test == 0, 0], X_test[y_test == 0, 1], color='#4285F4', marker='o', alpha=0.7, label='정상 (0)')
ax2.scatter(X_test[y_test == 1, 0], X_test[y_test == 1, 1], color='#EA4335', marker='s', alpha=0.7, label='위험/스팸 (1)')
ax2.set_title("테스트 데이터 (Test Data) 속성 분포")
ax2.set_xlabel("속성 1")
ax2.set_ylabel("속성 2")
ax2.legend()
ax2.grid(True, linestyle=':')

plt.tight_layout()
plt.show()
```

```python
# [실습 3] 4단계 모델링 구현 및 데이터 유형별(훈련 vs 테스트) 과대적합/과소적합 판별 시뮬레이션
from sklearn.neighbors import KNeighborsClassifier

# 데이터 학습에 과대적합(Overfitting)을 유도하기 위해 극단적으로 이웃의 수를 1로 설정한 모델과 일반 모델 대조
overfit_model = KNeighborsClassifier(n_neighbors=1)
good_model = KNeighborsClassifier(n_neighbors=11)

# 두 유형의 모델에 동일하게 훈련 데이터 모델링 적용
overfit_model.fit(X_train, y_train)
good_model.fit(X_train, y_train)

# 두 데이터 유형에 대한 정확도 성능 평가 비교 연산
train_acc_over = overfit_model.score(X_train, y_train)
test_acc_over = overfit_model.score(X_test, y_test)

train_acc_good = good_model.score(X_train, y_train)
test_acc_good = good_model.score(X_test, y_test)

print("=== 데이터 유형별 모델 성능 정량 대조 ===")
print("[모델 1: 과대적합 모델 (k=1)]")
print(f" - 훈련 데이터 정확도: {train_acc_over:.4f} / 테스트 데이터 정확도: {test_acc_over:.4f}")
print("[모델 2: 일반화 모델 (k=11)]")
print(f" - 훈련 데이터 정확도: {train_acc_good:.4f} / 테스트 데이터 정확도: {test_acc_good:.4f}")

# 성능 지표 대조 시각화
categories = ['과대적합 모델(훈련)', '과대적합 모델(테스트)', '일반화 모델(훈련)', '일반화 모델(테스트)']
accuracies = [train_acc_over, test_acc_over, train_acc_good, test_acc_good]

plt.figure(figsize=(10, 5))
colors = ['#1a73e8', '#aecbfa', '#e8eaed', '#bdc1c6']
bars = plt.bar(categories, accuracies, color=['#4285F4', '#9AA0A6', '#34A853', '#A8DADC'], width=0.4)
plt.ylim(0, 1.1)
plt.title("훈련 데이터와 테스트 데이터의 정확도 갭을 통한 과대적합 분석")
plt.ylabel("정확도 (Accuracy)")

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.02, f'{yval:.3f}', ha='center', va='bottom')

plt.show()
```

## 4. 혼동 행렬을 기반으로 한 최종 성능 평가 방법론

교재 182쪽에서 명시하는 분류 모델의 평가는 테스트 데이터를 통해 도출된 참(True)/거짓(False), 양성(Positive)/음성(Negative)의 4가지 조합 결합인 **혼동 행렬(Confusion Matrix)**을 기반으로 수행됩니다.

### **(1) 오차 행렬의 사상 구조**
* **TN (True Negative)**: 실제 정상(음성) 데이터를 모델이 정상(음성)으로 올바르게 분류한 수
* **FP (False Positive)**: 실제 정상(음성) 데이터를 모델이 스팸(양성)으로 잘못 분류한 수 (Type I Error)
* **FN (False Negative)**: 실제 스팸(양성) 데이터를 모델이 정상(음성)으로 잘못 분류한 수 (Type II Error)
* **TP (True Positive)**: 실제 스팸(양성) 데이터를 모델이 스팸(양성)으로 올바르게 분류한 수

### **(2) 수식 기반 정량 평가 지표**
모델의 최종 강건성은 테스트 데이터 세트 상에서 계산된 세 가지 지표를 통해 정의됩니다.

1. **정확도 (Accuracy)**: 전체 분할 테스트 데이터 중 모델이 올바르게 맞춘 클래스의 비율
   $$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$
2. **정밀도 (Precision)**: 모델이 Positive(스팸/질병)라고 예측한 총량 중 실제 Positive인 비율
   $$Precision = \frac{TP}{TP + FP}$$
3. **재현율 (Recall)**: 실제 데이터 속성 중 Positive인 전체 진본 중 모델이 Positive라고 유의미하게 검출해낸 비율
   $$Recall = \frac{TP}{TP + FN}$$

```python
# [실습 4] 테스트 데이터를 활용한 혼동 행렬 출력 및 최종 지표 검증
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# 일반화 모델(good_model)을 사용해 테스트 데이터의 예측 범주 도출
y_pred = good_model.predict(X_test)

# 혼동 행렬 연산
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)

print("=== 4단계: 테스트 데이터 기반 성능 평가 최종 리포트 ===")
print("Confusion Matrix:")
print(cm)
print(f"계산된 정확도 (Accuracy) : {acc:.4f}")
print(f"계산된 정밀도 (Precision): {prec:.4f}")
print(f"계산된 재현율 (Recall)   : {rec:.4f}")
```
