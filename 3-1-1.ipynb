## 2. 데이터 속성은 기계학습에 어떻게 사용될까?

* 데이터 속성에는 기계학습에 영향을 주는 속성과 영향을 받는 속성이 있다.

### **더 알아보기: 연속형 데이터와 범주형 데이터**

#### **[예시 3] 감기 증상을 알기 위한 데이터 속성 테이블**

| 환자 번호 | 체온(℃) | 기침 | 피로감 | 콧물 정도 (5점 척도) | 오한 정도 (5점 척도) | 감기 여부 |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| P01 | 37.5 | 있다 | 2 | 있다 | 5 | 감기 |
| P02 | 35.8 | 없다 | 1 | 없다 | 3 | 정상 |
| p03 | 36.2 | 없다 | 1 | 없다 | 1 | 정상 |
| P04 | 37.5 | 있다 | 2 | 있다 | 5 | 감기 |

* **연속형 데이터**: 체온의 37.5도와 같이 실수형 수치이고, 연속적으로 변화하는 값
* **범주형 데이터**: 기침이나 피로감처럼 '있다', '없다'와 같이 두 가지 상태 값으로 구분되는 데이터 및 정해진 문자 형태의 값
* **순서형 데이터**: 5점 척도와 같이 순서가 있는 범주형 데이터

### **기계학습에서의 역할 분류: 독립 변수와 종속 변수**

* **독립 변수**: 입력 값이나 원인을 나타내는 변수
* **종속 변수**: 결과나 효과를 나타내는 변수 (특성값에 영향을 받음)
* **기타**: 인과관계 분석에 직접 관여하지 않는 일련번호 등

### **[사례 분석] 로널드 피셔의 붓꽃(Iris) 데이터 세트 구조**
* 붓꽃의 품종과 꽃잎, 꽃받침의 길이와 너비 등을 센티미터 단위로 측정한 150개의 데이터 집합으로 구성
* 품종: `Iris-setosa`, `Iris-versicolor`, `Iris-virginica`

#### **붓꽃 데이터 속성 상세 테이블**

| 속성명 | Id | Sepal Length | Sepal Width | Petal Length | Petal Width | Species |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| **의미** | 일련번호 | 꽃받침 길이 | 꽃받침 너비 | 꽃잎 길이 | 꽃잎 너비 | 품종 |
| **속성값** | 1 | 5.1 | 3.5 | 1.4 | 0.2 | Iris-setosa |
| **데이터 타입** | 수치형 | 수치형 | 수치형 | 수치형 | 수치형 | 범주형 |
| **역할** | 기타 | 독립 변수 | 독립 변수 | 독립 변수 | 독립 변수 | 종속 변수 |

#### **붓꽃 데이터 실제 샘플 예시**

| Id | Sepal Length | Sepal Width | Petal Length | Petal Width | Species |
| :--- | :--- | :--- | :--- | :--- | :--- |
| 1 | 5.1 | 3.5 | 1.4 | 0.2 | Iris-setosa |
| 2 | 4.9 | 3.0 | 1.4 | 0.2 | Iris-setosa |
| 51 | 7.0 | 3.2 | 4.7 | 1.4 | Iris-versicolor |
| 52 | 6.4 | 3.2 | 4.5 | 1.5 | Iris-versicolor |
| 101 | 6.3 | 3.3 | 6.0 | 2.5 | Iris-virginica |
| 102 | 5.8 | 2.7 | 5.1 | 1.9 | Iris-virginica |

#### **품종별 데이터 속성값 특징 비교**
* **꽃받침**: 세 가지 품종의 꽃받침의 길이와 너비는 큰 차이가 없다.
* **꽃잎**: 품종 결정에 결정적인 영향을 줌
  * `Iris-setosa`: 세 가지 품종 중 꽃잎의 길이와 꽃잎의 너비가 가장 짧고 작다.
  * `Iris-versicolor`: 세 가지 품종 중 꽃잎의 길이와 꽃잎의 너비가 각각 중간이다.
  * `Iris-virginica`: 세 가지 품종 중 꽃잎의 길이와 꽃잎의 너비가 가장 길고 크다.

```python
# 실습: 기계학습을 위한 데이터 역할 분리 (독립변수 X 와 종속변수 y)
import pandas as pd

# 1. 교재의 붓꽃 데이터 수집 및 데이터프레임 생성
iris_data = {
    "Id": [1, 2, 51, 52, 101, 102],
    "SepalLength": [5.1, 4.9, 7.0, 6.4, 6.3, 5.8],
    "SepalWidth": [3.5, 3.0, 3.2, 3.2, 3.3, 2.7],
    "PetalLength": [1.4, 1.4, 4.7, 4.5, 6.0, 5.1],
    "PetalWidth": [0.2, 0.2, 1.4, 1.5, 2.5, 1.9],
    "Species": ["Iris-setosa", "Iris-setosa", "Iris-versicolor", "Iris-versicolor", "Iris-virginica", "Iris-virginica"]
}
df_iris = pd.DataFrame(iris_data)

# 2. 독립 변수 (X) 추출: 원인에 해당하며 품종 결정에 영향을 주는 속성들
X_iris = df_iris[["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]]
print("=== [독립 변수 (입력 특성 - Features)] ===")
display(X_iris)

# 3. 종속 변수 (y) 추출: 결과에 해당하며 독립변수의 특성값에 영향을 받는 속성
y_iris = df_iris["Species"]
print("\n=== [종속 변수 (결과/라벨 - Target Label)] ===")
print(y_iris)

print("\n" + "="*50 + "\n")

# 4. 연계 실습: K-pop 아이돌 데이터에서 독립 변수와 종속 변수 분리하기
# 소속사와 멤버 수, 데뷔 연도를 원인(독립 변수)으로 두고, 대표곡을 결과(종속 변수)로 분리하는 예제
X_idols = df_idols[["멤버 수", "소속사", "데뷔 연도"]]
y_idols = df_idols["대표곡"]

print("=== [K-pop 아이돌 데이터 독립 변수] ===")
display(X_idols)
print("\n=== [K-pop 아이돌 데이터 종속 변수] ===")
print(y_idols)
```

## 수치형 데이터와 범주형 데이터 구분하기

K-pop 아이돌 데이터에서도 수치형 데이터와 범주형 데이터를 구분할 수 있다.

수치형 데이터는 숫자로 표현되며 크기 비교나 평균 계산이 가능한 데이터이다.

범주형 데이터는 정해진 범주나 이름으로 구분되는 데이터이다.

| 데이터 유형 | 속성 예시 | 설명 |
|---|---|---|
| 수치형 데이터 | Height, Weight | 키와 몸무게처럼 숫자로 표현되는 값 |
| 범주형 데이터 | Group, Company, Country, Gender | 그룹, 소속사, 국가, 성별처럼 범주로 구분되는 값 |

단, 생년월일이나 데뷔일은 숫자가 포함되어 있지만 단순한 수치라기보다 날짜형 데이터로 다루는 것이 적절하다.

```python
# [실습] 수치형 데이터와 범주형 데이터 나누어 보기

numeric_columns = ["Height", "Weight"]
categorical_columns = ["Stage Name", "Group", "Company", "Country", "Gender"]

print("=== 수치형 데이터 예시 ===")
display(df_kpop[numeric_columns].head())

print("\n=== 범주형 데이터 예시 ===")
display(df_kpop[categorical_columns].head())
```

## 결측값 확인하기

외부에서 수집한 실제 데이터에는 빈 값이 포함될 수 있다.

이처럼 값이 비어 있는 데이터를 결측값이라고 한다.

예를 들어 어떤 아이돌의 키나 몸무게 정보가 공개되어 있지 않다면 해당 칸은 비어 있을 수 있다.

기계학습에서는 결측값을 그대로 사용할 수 없는 경우가 많기 때문에, 데이터 분석 전에 결측값을 확인하는 과정이 필요하다.

```python
# [실습] 결측값 확인하기

missing_values = df_kpop.isna().sum().sort_values(ascending=False)

print("=== 속성별 결측값 개수 ===")
display(missing_values)
```

## 그룹별 아이돌 수 확인하기

범주형 데이터는 같은 범주에 속하는 데이터가 몇 개 있는지 세어볼 수 있다.

예를 들어 `Group` 속성을 이용하면 그룹별로 몇 명의 아이돌이 데이터에 포함되어 있는지 확인할 수 있다.

이 과정은 범주형 데이터의 분포를 이해하는 데 도움이 된다.

```python
# [실습] 그룹별 아이돌 수 확인하기

group_counts = df_kpop["Group"].value_counts()

print("=== 그룹별 아이돌 수 상위 20개 ===")
display(group_counts.head(20))
```
